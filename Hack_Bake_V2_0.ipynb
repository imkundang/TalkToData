```python
df1= pd.read_csv('/content/Fund_data_final.csv')
```

```python
df1.columns
```

```python
df_source2 = df1.drop(['Unnamed: 0'],axis=1)
```

```python
df_source2 = df_source2[['FundID', 'Fund_Name']]
```

```python
df_source2['Fund_Name'].unique()
```

```python
len(df_source2['Fund_Name'].unique())
```

```python
import random
num_ids_to_modify = 5
ids_to_modify = random.sample(df_source2['FundID'].tolist(), min(num_ids_to_modify, len(df_source2)))

# Step 2: Modify fund names for the selected IDs
new_fund_names = ['Daiwa ETF TOPIX V', 'Nomura TOPIX ETF XI','VanEck Vectors ChinaAMC VI',
       'First Trust Emerging Markets AI Theme',
       'KraneShares E Fund Tech VI' ]


```

```python
# Update the fund names for the selected IDs
for i, fund_id in enumerate(ids_to_modify):
    df_source2.loc[df_source2['FundID'] == fund_id, 'Fund_Name'] = new_fund_names[i]

# Display the modified DataFrame
print("\nModified DataFrame:")
df_source2.head()
```

```python
df_source2.rename(columns = {'Fund_Name':'Fund_Name_Modified'}, inplace = True)
```

```python
df_join = pd.merge(df1,df_source2,on='FundID',how='left')
```

```python
df_join.to_csv('merged.csv')
```

```python
df_join.head()
```

```python
mismatched_df = df_join[df_join['Fund_Name'] != df_join['Fund_Name_Modified']]
```

```python
mismatched_df  = mismatched_df.drop(['Unnamed: 0'],axis=1)
```
